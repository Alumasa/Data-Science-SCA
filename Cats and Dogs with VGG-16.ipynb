- Import the required libraries from tensorflow.

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dense, Flatten, BatchNormalization, Conv2D, MaxPool2D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import random
import numpy as np
from sklearn.metrics import confusion_matrix
import itertools
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
%matplotlib inline
```

- Place the data paths into variables.

```python
train_path = 'datasets/training_set'
test_path = 'datasets/test_set'
```

- Preprocess the test and train images using the vgg16 application in Keras.

```python
train_batches = ImageDataGenerator(preprocessing_function=tf.keras.applications.vgg16.preprocess_input)\
                .flow_from_directory(directory=train_path, target_size=(224, 224), classes=['cats', 'dogs'], batch_size=10)
test_batches = ImageDataGenerator(preprocessing_function=tf.keras.applications.vgg16.preprocess_input)\
                .flow_from_directory(directory=test_path, target_size=(224, 224), classes=['cats', 'dogs'], batch_size=10, 
                                     shuffle=False)
```

- A helper function that prints a batch of images and their classification.
- [1. 0.] is cat and [0. 1.] is dog.

```python
def plotImages(images_arr):
    fig, axes = plt.subplots(1, 10, figsize=(20,20))
    axes = axes.flatten()
    for img, ax in zip( images_arr, axes):
        ax.imshow(img)
        ax.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
imgs, labels = next(train_batches)
plotImages(imgs)
print(labels)
```

- Define the VGG-16 model and look at its summary.

```python
vgg16_model = tf.keras.applications.vgg16.VGG16()
```

```python
vgg16_model.summary()
```

- Origially the VGG-16 model is pretrained to  give 1000 outputs; as can be seen on the predictions layer.
- I will select all the layers except the last one and add them to another model variable.
- In the new model varaible I will add the output layer with 2 neurons.

```python
model = Sequential()
for layer in vgg16_model.layers[:-1]:
    model.add(layer)
```

```python
for layer in model.layers:
    layer.trainable = False
```

```python
model.add(Dense(units=2, activation='softmax'))
```

```python
model.summary()
```

- Tune the network with an optimizer, loss function and metric.

```python
model.compile(optimizer=Adam(learning_rate=0.0001), loss='categorical_crossentropy', metrics=['accuracy'])
```

- Train the model.

```python
%%time
model.fit(x = train_batches, 
          steps_per_epoch = len(train_batches),
          epochs = 1,
          verbose = 2
         )
```

- Predict the test data.

```python
%%time
predictions = model.predict(x = test_batches, steps = len(test_batches))
```

- Plot a confusion matrix to visualize the predictions.

```python
pred_matrix = confusion_matrix(y_true=test_batches.classes, y_pred=np.argmax(predictions, axis=-1))
pred_matrix
```
